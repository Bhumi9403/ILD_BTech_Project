```python
import numpy as np
import cv2
```

```python
image = cv2.imread("Y:\\Dataset\\CASIA Iris Image Database (version 1.0)\\003\\1\\003_1_2.bmp",0)
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
otsu_threshold, images_result = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print("Obtained threshold: ", otsu_threshold)
```

```python
image
```

```python
original_image = image.ravel()
otsu_image = images_result.ravel()
```

```python
print(original_image)
print(otsu_image)
```

```python
lower_half = []
upper_half = []
thresh = otsu_threshold

for x in original_image:
    if(x >= thresh):
        upper_half.append(x)
    else:
        lower_half.append(x)
        

print(lower_half)
print(upper_half)

```

```python
otsu_upper_avg = np.mean(upper_half)
otsu_lower_avg = np.mean(lower_half)

print(otsu_upper_avg)
print(otsu_lower_avg)
```
