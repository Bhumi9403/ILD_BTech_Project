```python
import numpy as np
import pandas as pd
import cv2
import os
```

```python
r1 = []
r2 = []
g1 = []
g2 = []
b1 = []
b2 = []

images = []
imgname=[]
def tsbtc_2arry(folder):
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
        imgname.append(filename)
        red_c = img[:,:,2]
        green_c = img[:,:,1]
        blue_c = img[:,:,0]
        
        red_arr = red_c.ravel()
        green_arr = green_c.ravel()
        blue_arr = blue_c.ravel()
        
        red_arr.sort()
        green_arr.sort()
        blue_arr.sort()
        
        red_split = np.array_split(red_arr,2)
        green_split = np.array_split(green_arr,2)
        blue_split = np.array_split(blue_arr,2)
        
        r1.append(np.average(red_split[0]))
        r2.append(np.average(red_split[1]))
        
        g1.append(np.average(green_split[0]))
        g2.append(np.average(green_split[1]))
        
        b1.append(np.average(blue_split[0]))
        b2.append(np.average(blue_split[1]))
        
        
        
        
        
        
```

```python
tsbtc_2arry('Y:\Wallpapers\Assassians Creed')
```
